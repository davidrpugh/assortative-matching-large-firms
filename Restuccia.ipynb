# Increasing variance of $x$ using the parameters from Adamopoulos and Restuccia
## 1. Rich country - constant $x$
The next cell contains the distribution and density for workers and firms.

```python
# define some workers skill
x, a, b = sym.var('x, a, b')
skill_cdf = (x - a) / (b - a)
skill_params = {'a': 0.9999, 'b': 1.0001}
skill_bounds = [skill_params['a'], skill_params['b']]

workers = inputs.Input(var=x,
                       cdf=skill_cdf,
                       params=skill_params,
                       bounds=skill_bounds,
                       )

# define some firms
y, mu2, sigma2 = sym.var('y, mu2, sigma2')
productivity_cdf = 0.5 + 0.5 * sym.erf((sym.log(y) - mu2) / sym.sqrt(2 * sigma2**2))
productivity_params = {'mu2': -1.83, 'sigma2': 4.66}
productivity_bounds = [1e-3, 5e1]

firms = inputs.Input(var=y,
                     cdf=productivity_cdf,
                     params=productivity_params,
                     bounds=productivity_bounds,
                     )
```

This next one declares the necessary variables and parameters (for a rich country) and the capital function from the notebook *Capital* 

```python
# Developped country

R, l, r, rho, gamma, eta, L, A, kapa = sym.var('R, l, r, rho, gamma, eta, L, A, kapa')

# First set k to root 3
k_star = (2*6**(-0.5)*(A*eta**2*kapa*x**0.5/R)**0.5*sym.cos(0.333333333333333*sym.acos(3*6**0.5*x**(-0.5)*(l*x*y/r)**0.25*(R*x**(-0.5)/(A*eta**2*kapa))**0.5*(-eta + 1)/(2*eta))))**(4)

F = A*kapa*(eta*(k_star*x)**rho + (1- eta)*(y*(l/r))**rho)**(gamma/rho)

F_params = {'R':0.13099, 'rho':0.25, 'gamma':0.5, 'eta':0.89, 'L':4.2, 'A': 1.0, 'kapa':1.0}
```

Checking capital is postive...

```python
kk = sym.lambdify((x,y,l,r,A,kapa,eta,R), k_star)
kk(0.75,1,25,1,1,1,0.89,0.13099)
```

These next two cells introduce the production function with the optimal capital in the solver, and solve it.

```python
# Try out positive
model = models.Model('positive',
                     workers=workers,
                     firms=firms,
                     production=F,
                     params=F_params)

solver = solvers.ShootingSolver(model=model)
```

```python
solver.solve(2.0, tol=1e-6, number_knots=100, integrator='lsoda',
             with_jacobian=True, atol=1e-12, rtol=1e-9, check=False)
```

### This next cell tells you the solution for PAM of the model!

```python
solver.solution
```

These two next cells extract data from the solution and plot it - incluiding capital, using the numerical version of $k^{*}(x,y,l)$

```python
xs = []
thetas = []
ys = []
ws = []
for i in range(100):
    xs.append(solver.solution.index[i])
    ys.append(solver.solution.iloc[i,0])
    thetas.append(solver.solution.iloc[i,1])
    ws.append(solver.solution.iloc[i,2])
ks = []
for a in range(len(xs)):
    ks.append(kk(xs[a],ys[a],thetas[a],1,1,1,0.89,0.13099))
kk(xs[95],ys[95],thetas[95],1,1,1,0.89,0.13099), kk(xs[0],ys[0],thetas[0],1,1,1,0.89,0.13099)
```

```python
plt.figure(1, figsize=(16, 4))
plt.suptitle("Rich Country, no variation in x", fontsize = 15)

plt.subplot(131)
plt.xlabel("$y$", fontsize = 20)
plt.ylabel("$k^{*},x$", fontsize=20)
plt.plot(ys, xs, label="Land quality")
plt.legend(loc=2)
plt.xlim(0,0.08)
plt.ylim(0,2)

plt.subplot(132)
plt.plot(ys, thetas, label="Labour", c='r')
plt.xlabel("$y$", fontsize = 20)
plt.legend(loc=2)
plt.xlim(0,0.08)
#plt.ylim(15,18)

plt.subplot(133)
plt.plot(ys, ks, label="Capital", c='g')
plt.xlabel("$y$", fontsize = 20)
plt.legend(loc=2)
plt.xlim(1,0.08)
plt.ylim(8,10)

plt.show()
```

```python
Rs = np.ones(len(ys))
Rs = Rs*0.13099
plt.plot(ys, ws, label="Land rental price", c='g')
#plt.plot(ys, Rs, label="Capital rental price", c='b')
plt.xlabel("$y$", fontsize = 20)
plt.ylabel("$w$", fontsize = 20)
plt.legend(loc='best')
plt.xlim(0,0.08)
plt.ylim(0,1)

plt.show()
```
