```python
import matplotlib.pyplot as plt
from scipy import stats, optimize
import numpy as np

from __future__ import division
```

```python
#Plotting Log-Normal with mean=0 and var=1 for different location parameters
mu = 0.0
sigma = 1.0
loc1 = 0.0
loc2 = -1e-1
loc3 = 1.0
x1 = np.linspace(loc1+1e-3,5.0,1000)
x2 = np.linspace(loc2+1e-3,5.0,1000)
x3 = np.linspace(loc3+1e-3,5.0,1000)
```

```python
#s=sigma, loc=shifts entire distribution,scale=exp(mu)
cdf1 = stats.lognorm.cdf(x1,s=1.0,loc=loc1,scale=1.0)
cdf2 = stats.lognorm.cdf(x2,s=1.0,loc=loc2,scale=1.0)
cdf3 = stats.lognorm.cdf(x3,s=1.0,loc=loc3,scale=1.0)
```

```python
plt.plot(x1,cdf1)
plt.plot(x2,cdf2)
plt.plot(x3,cdf3)
plt.title("CDF")
plt.show()
```

```python
pdf1 = stats.lognorm.pdf(x1,s=1.0,loc=loc1,scale=1.0)
pdf2 = stats.lognorm.pdf(x2,s=1.0,loc=loc2,scale=1.0)
pdf3 = stats.lognorm.pdf(x3,s=1.0,loc=loc3,scale=1.0)

plt.plot(x1,pdf1)
plt.plot(x2,pdf2)
plt.plot(x3,pdf3)
plt.title("PDF")
plt.show()
```
