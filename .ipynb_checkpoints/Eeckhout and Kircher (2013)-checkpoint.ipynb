## Multiplicative separability
A particularly attractive case arises under multiplicative separability of the form

$$F(x, y, l, r) = A(x, y)B(l, r)$$

In this case the condition for positive assortative matching can be written as 

$$\frac{AA_{xy}}{A_xA_y}\frac{BB_{lr}}{B_lB_r} \ge 1$$

If $B$ has constant elasticity of substitution form, then we obtain an even simpler condition

$$\frac{AA_{xy}}{A_xA_y} \frac{1}{\sigma_{lr}}\ge 1$$

where $\sigma_{lr}$ is the elasticity of substitution between $l$ and $r$. 

Finally, if $A$ also has constant elasticity of substitution form, then we obtain an even simpler condition

$$\frac{1}{\sigma_{xy}\sigma_{lr}}\ge 1 \iff \sigma_{xy}\sigma_{lr} \le 1$$

where $\sigma_{xy}$ is the elasticity of substitution between $x$ and $y$.

```python
def A(x, y, params):
    """
    Inner production function.

    Arugments:

        x: Worker type.
        y: Firm type.

    """ 
    sigma_A = params['sigma_A']
    omega_A = params['omega_A']
    
    if sigma_A == 1.0:
        out = x**omega_A * y**(1 - omega_A)
    else:
        out = ((omega_A * x**((sigma_A - 1) / sigma_A) + 
               (1 - omega_A) * y**((sigma_A - 1) / sigma_A))**(sigma_A / (sigma_A - 1))) 
    return out
 
def B(l, r, params):
    """
    Inner production function.
    
    Arguments:

        l: Quantity of labor of type x.
        r: Fraction of proprietary resources devoted to a worker
           with type x (can be normalized to unity for convenience).

    """ 
    sigma_B = params['sigma_B']
    omega_B = params['omega_B']
    
    if sigma_B == 1.0:
        out = l**omega_B * r**(1 - omega_B)
    else:
        out = ((omega_B * l**((sigma_B - 1) / sigma_B) + 
               (1 - omega_B) * r**((sigma_B - 1) / sigma_B))**(sigma_B / (sigma_B - 1)))
    return out

def F(x, y, l, r, params):
    """
    Nested CES version of production technology.

    Arguments:

        x:      worker type
        y:      firm type
        l:      quantity of labor of type x
        r:      fraction of proprietary resources devoted to a worker
                with type x (can be normalized to unity for convenience)
  
    The first two arguments x and y are quality variables describing the worker 
    and firm types, while the latter two aruments are quantity variables 
    describing the level of inputs.

    """
    return A(x, y, params) * B(l, r, params)
```

# Example with log-normal workers and firms.

```python
# distribution of worker types
mu, sigma = 0.5, 0.5
worker_types = stats.lognorm(sigma, scale=np.exp(mu))
eps = 1e-4
x_lower, x_upper, x_measure = worker_types.ppf(eps), worker_types.ppf(1-eps), 1.0
workers = eeckhoutKircher2013.Workers(worker_types, x_lower, x_upper, x_measure)

# distribution of firm productivity
firm_prods = stats.lognorm(sigma, scale=np.exp(mu)) 
y_lower, y_upper, y_measure = firm_prods.ppf(eps), firm_prods.ppf(1-eps), 1.0
firms = eeckhoutKircher2013.Firms(firm_prods, y_lower, y_upper, y_measure)
```

```python
# model parameters
params = {'omega_A':0.5, 'omega_B':0.45, 'sigma_A':2.0, 'sigma_B':1.0}

# create an object representing the model
model = eeckhoutKircher2013.Model(params, workers, firms, F)

# degree of polynomial approximation
deg = 3

# initial guess for theta(x) is constant function
init_theta_coefs    = np.zeros(deg + 1)
init_theta_coefs[0] = model.firms.size / model.workers.size

# initial guess for mu(x) is identity function
y_lower, y_upper = model.firms.lower_bound, model.firms.upper_bound
init_mu_coefs = np.zeros(deg + 1)
init_mu_coefs[0] = 0.5 * (y_upper + y_lower)
init_mu_coefs[1] = 0.5 * (y_upper - y_lower)
        
# initial set of coefficients should have shape (2, deg + 1)
init_coefs = np.vstack((init_theta_coefs, init_mu_coefs))
tmp_coefs = init_coefs

# define the boundary conditions...should be (x, mu(x)) pairs!
boundary_conditions = [(model.workers.lower_bound, model.firms.lower_bound),
                       (model.workers.upper_bound, model.firms.upper_bound)]

    
# compute the equilibrium using collocation
model.set_solver('OrthogonalCollocation', kind='Chebyshev')
result = model.solver.solve(init_coefs, bc=boundary_conditions, matching='pam', 
                            method='hybr', options={'full_output':True})
    
```

```python
def monotonicity_boundary(x, x_upper, y_upper, params):
    """Matching functions that cross this boundary lead to non-monotonic theta(y)."""
    omega_A = params['omega_A']
    omega_B = params['omega_B']
    sigma_A = params['sigma_A']
    
    A = (omega_A / omega_B) * ((1 - omega_B) / (1 - omega_A)) 
    B = (x**((sigma_A - 1) / sigma_A) - x_upper**((sigma_A - 1) / sigma_A))
    C = y_upper**((sigma_A - 1) / sigma_A)
    
    return (A * B + C)**(sigma_A / (sigma_A - 1))
```

```python
def boundary_condition_1(x, params):
    # extract params
    omega_A = params['omega_A']
    sigma_A = params['sigma_A']
    
    numerator   = omega_A * (1 - omega_A) * ((1 / (1 - sigma_A)) + 1) + (1-omega_A) * (1 - 2 * omega_A)
    denominator = omega_A * (1 - omega_A) * ((1 / (1 - sigma_A)) + 1) - omega_A * (1 - 2 * omega_A)
    
    return (numerator / denominator)**(sigma_A / (1 - sigma_A)) * x

def boundary_condition_2(x, params):
    # extract params
    omega_A = params['omega_A']
    sigma_A = params['sigma_A']
    
    numerator = (1 - omega_A) * (1 / (1 - sigma_A)) + omega_A - 1  
    denominator = (1 - omega_A) * (1 / (1 - sigma_A)) + omega_A
    
    return (numerator / denominator)**(sigma_A / (1 - sigma_A)) * x
```

## Multiplicative Separability, Redux!

A slight twist on the multiplicative separability functional form above...

$$F(x, y, l, r) = A(x, y)B(y, l, r)$$

with 

$$A(x, y) = \bigg[\omega_A x^{\rho_A} + (1 - \omega_A)y^{\rho_A}\bigg]^{\frac{1}{\rho_A}}$$

and

$$B(y, l, r) = r \left[\omega_B y^{\rho_B} + (1 - \omega_B)\left(\frac{l}{r}\right)^{\rho_B}\right]^{\frac{1}{\rho_B}}$$

```python
def A(x, y, params):
    """Inner production function.""" 
    omega_A = params['omega_A']
    rho_A   = params['rho_A']
    
    return (omega_A * x**rho_A + (1 - omega_A) * y**rho_A)**(1 / rho_A)
 
def B(y, l, r, params):
    """Inner production function.""" 
    omega_B = params['omega_B']
    rho_B   = params['rho_B']
    
    return r * (omega_B * y**rho_B + (1 - omega_B) * (l / r)**rho_B)**(1 / rho_B)

def F(x, y, l, r, params):
    """Nested CES version of production technology.

    Arguments:

        x:      worker type
        y:      firm type
        l:      quantity of labor of type x
        r:      fraction of proprietary resources devoted to a worker
                with type x (can be normalized to unity for convenience)
  
    The first two arguments x and y are quality variables describing 
    the worker and firm types, while the latter two aruments are 
    quantity variables describing the level of inputs.

    """
    return A(x, y, params) * B(y, l, r, params)
```

```python
# model parameters
params = {'omega_A':0.95, 'omega_B':0.25, 'rho_A':2.0, 'rho_B':2.0}

# distribution of worker types
x_lower, x_upper = 1, 5
workerTypes = stats.uniform(x_lower, x_upper - x_lower)
workers = (workerTypes, x_lower, x_upper)

# distribution of firm productivity
y_lower, y_upper = 1, 5
firmProds = stats.uniform(y_lower, y_upper - y_lower)
firms = (firmProds, y_lower, y_upper)

# create an object representing the model
model = eeckhoutKircher2013.Model(params, workers, firms, F)
```

## Parameters and original production function from Eeckhout and Kircher (2013)

```python
def F(x, y, l, r, params):
    """Nested CES version of production technology.

    Arguments:

        x:      worker type
        y:      firm type
        l:      quantity of labor of type x
        r:      fraction of proprietary resources devoted to a worker
                with type x (can be normalized to unity for convenience)
  
    The first two arguments x and y are quality variables describing 
    the worker and firm types, while the latter two aruments are 
    quantity variables describing the level of inputs.

    """
    # extract params
    alpha   = params['alpha']
    beta    = params['beta']
    gamma   = params['gamma']
    lambda1 = params['lambda1']
    lambda2 = params['lambda2']
    mu      = params['mu']
    
    out = r * (mu * y**alpha + (1 - mu) * (lambda1 * x**beta + lambda2 * (l / r)**beta)**(alpha / beta))**gamma
    return out
```

```python
def F2(x, y, l, r, params):
    """Limiting case of the production function F for beta=0 and 
    lambda1=lambda2=1. Skills remain heterogenous but the distribution
    is indeterminate since skills are perfect substitutes.

    Arguments:

        x:      worker type
        y:      firm type
        l:      quantity of labor of type x
        r:      fraction of proprietary resources devoted to a worker
                with type x (can be normalized to unity for convenience)
  
    The first two arguments x and y are quality variables describing 
    the worker and firm types, while the latter two aruments are 
    quantity variables describing the level of inputs.

    """
    # extract params
    alpha   = params['alpha']
    gamma   = params['gamma']
    mu      = params['mu']
    
    out = r * (mu * y**alpha + (1 - mu) * (x * (l / r))**alpha)**gamma
    return out
```

```python
def F3(x, y, l, r, params):
    """Limiting case of the production function F for alpha=1/gamma=0,
    and r=1.  In this case the model is basically the Lucas span of 
    control model in efficiency units xl.

    Arguments:

        x:      worker type
        y:      firm type
        l:      quantity of labor of type x
        r:      fraction of proprietary resources devoted to a worker
                with type x (can be normalized to unity for convenience)
  
    The first two arguments x and y are quality variables describing 
    the worker and firm types, while the latter two aruments are 
    quantity variables describing the level of inputs.

    """
    # extract params
    mu      = params['mu']
    
    out = y**mu * (x * l)**(1 - mu)
    return out
```

## Spectral methods: Orthogonal Collocation

```python
# describe the workers
mu, sigma = 0, 1
worker_types = stats.lognorm(sigma, 0.0, np.exp(mu)) 
x_lower, x_upper, x_measure = worker_types.ppf(0.01), worker_types.ppf(0.99), 1
workers = eeckhoutKircher2013.Workers(worker_types, x_lower, x_upper, x_measure)

# describe the firms
firm_prods = stats.lognorm(sigma, 0.0, np.exp(mu)) 
y_lower, y_upper, y_measure = firm_prods.ppf(0.01), firm_prods.ppf(0.99), 1
firms = eeckhoutKircher2013.Firms(firm_prods, y_lower, y_upper, y_measure)

# initialize parameters
params = {'omega_A':0.5, 'omega_B':0.45, 'sigma_A':0.25, 'sigma_B':1.0}

# initialize the model
model = eeckhoutKircher2013.Model(params, workers, firms, F)
```

```python
# define the boundary conditions...should be (x, mu(x)) pairs!
boundary_conditions = [(model.workers.lower_bound, model.firms.lower_bound),
                       (model.workers.upper_bound, model.firms.upper_bound)]
```

```python
# degree of polynomial approximation
deg = 75

# initial guess for theta(x) is constant function
init_theta_coefs    = np.zeros(deg + 1)
init_theta_coefs[0] = model.workers.size
        
# initial guess for mu(x) is identity function (coefficients depend on basis funcs!)
y_lower, y_upper = model.firms.lower_bound, model.firms.upper_bound
init_mu_coefs = np.zeros(deg + 1)
init_mu_coefs[0] = 0.5 * (y_upper + y_lower)
init_mu_coefs[1] = 0.5 * (y_upper - y_lower)
        
# initial set of coefficients should have shape (2, deg + 1)
init_coefs = np.vstack((init_theta_coefs, init_mu_coefs))
```

```python
# compute the equilibrium using collocation
model.set_solver('OrthogonalCollocation', kind='Chebyshev')
model.solver.solve(init_coefs, bc=boundary_conditions, matching='pam', method='hybr')
```
