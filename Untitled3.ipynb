```python
import numpy as np
from scipy import interpolate, stats
import sympy as sp
from sympy.interactive import printing
printing.init_printing(use_latex=True)

import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

import eeckhoutKircher2013
```

```python
def A(x, y, params):
    """
    Inner production function.

    Arugments:

        x: Worker type.
        y: Firm type.

    """ 
    sigma_A = params['sigma_A']
    omega_A = params['omega_A']
    
    if sigma_A == 1.0:
        out = x**omega_A * y**(1 - omega_A)
    else:
        out = ((omega_A * x**((sigma_A - 1) / sigma_A) + 
               (1 - omega_A) * y**((sigma_A - 1) / sigma_A))**(sigma_A / (sigma_A - 1))) 
    return out
 
def B(l, r, params):
    """
    Inner production function.
    
    Arguments:

        l: Quantity of labor of type x.
        r: Fraction of proprietary resources devoted to a worker
           with type x (can be normalized to unity for convenience).

    """ 
    sigma_B = params['sigma_B']
    omega_B = params['omega_B']
    
    if sigma_B == 1.0:
        out = l**omega_B * r**(1 - omega_B)
    else:
        out = ((omega_B * l**((sigma_B - 1) / sigma_B) + 
               (1 - omega_B) * r**((sigma_B - 1) / sigma_B))**(sigma_B / (sigma_B - 1)))
    return out

def F(x, y, l, r, params):
    """
    Nested CES version of production technology.

    Arguments:

        x:      worker type
        y:      firm type
        l:      quantity of labor of type x
        r:      fraction of proprietary resources devoted to a worker
                with type x (can be normalized to unity for convenience)
  
    The first two arguments x and y are quality variables describing the worker 
    and firm types, while the latter two aruments are quantity variables 
    describing the level of inputs.

    """
    return A(x, y, params) * B(l, r, params)
```

```python
# distribution of worker types
mu, sigma = 0, 1
worker_types = stats.lognorm(sigma, scale=np.exp(mu)) 
x_lower, x_upper, x_measure = worker_types.ppf(0.01), worker_types.ppf(0.99), 1
workers = eeckhoutKircher2013.Workers(worker_types, x_lower, x_upper, x_measure)

# distribution of firm productivity
firm_prods = stats.lognorm(sigma, scale=np.exp(mu)) 
y_lower, y_upper, y_measure = firm_prods.ppf(0.01), firm_prods.ppf(0.99), 1
firms = eeckhoutKircher2013.Firms(firm_prods, y_lower, y_upper, y_measure)
```

```python
# define symbolic variables
x, y, l, r = sp.symbols('x,y,l,r')

# define symbolic parameters
omega_A, sigma_A, omega_B, sigma_B = sp.symbols('omega_A,sigma_A,omega_B,sigma_B')

rho_A      = (sigma_A - 1) / sigma_A 
symbolic_A = (omega_A * x**rho_A + (1 - omega_A) * y**rho_A)**(1 / rho_A) 
symbolic_B = l**omega_B * r **(1 - omega_B)
symbolic_F = symbolic_A * symbolic_B 
```

Compute the partial and cross partial derivatives

```python
Fx = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, x), modules='numpy')
Fy = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, y), modules='numpy')
Fl = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, l), modules='numpy')
Fxy = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, x, y), modules='numpy')
Fyl = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, y, l), modules='numpy')
Fxl = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, x, l), modules='numpy')
Fxr = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, x, r), modules='numpy')
Fyr = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, y, r), modules='numpy')
Flr = sp.lambdify((x,y,l,r,omega_A,sigma_A, omega_B), sp.diff(symbolic_F, l, r), modules='numpy')
```

```python
# initialize parameters
params = {'omega_A':0.5, 'omega_B':0.5, 'sigma_A':0.5, 'sigma_B':1.0}

# initialize the model
model = eeckhoutKircher2013.Model(params, workers, firms, F)

# degree of polynomial approximation
deg = 95

# initial guess for theta(x) is constant function
init_theta_coefs    = np.zeros(deg + 1)
init_theta_coefs[0] = model.workers.size / model.firms.size
        
# initial guess for mu(x) is identity function
y_lower, y_upper = model.firms.lower_bound, model.firms.upper_bound
init_mu_coefs = np.zeros(deg + 1)
init_mu_coefs[0] = 0.5 * (y_upper + y_lower)
init_mu_coefs[1] = 0.5 * (y_upper - y_lower)
        
# initial set of coefficients should have shape (2, deg + 1)
init_coefs = np.vstack((init_theta_coefs, init_mu_coefs))
tmp_coefs = init_coefs

# define the boundary conditions...should be (x, mu(x)) pairs!
boundary_conditions = [(model.workers.lower_bound, model.firms.lower_bound),
                       (model.workers.upper_bound, model.firms.upper_bound)]
```

```python
colors = mpl.cm.jet(np.linspace(0, 1, 50))

fig, ax = plt.subplots(1,2, figsize=(12,8), sharey=True)

param_values = np.linspace(0.5, 0.9, 25)
for i, val in enumerate(param_values):
    
    # change the parameter
    model.update_parameter('omega_A', val)
    
    # compute the equilibrium using collocation
    model.set_solver('OrthogonalCollocation', kind='Chebyshev')
    model.solver.solve(init_coefs, bc=boundary_conditions, matching='pam', method='hybr')
    
    tmp_theta = model.equilibrium['theta']
    tmp_mu    = model.equilibrium['mu']
    xgrid     = np.linspace(x_lower, x_upper, 1000)
    
    # mu(x) may be increasing or decreasing!
    data = np.hstack((xgrid[:,np.newaxis], tmp_mu(xgrid)[:,np.newaxis]))
    data = data[data[:,1].argsort()]
    inverse_mu = interpolate.PchipInterpolator(data[:,1], data[:,0])
        
    # grid of values for plotting the matching function
    mu_lower = tmp_mu(x_lower)
    mu_upper = tmp_mu(x_upper)
    mu_grid  = np.linspace(mu_lower, mu_upper, 1000)    

    marginal_product_skill = Fx(inverse_mu(mu_grid), mu_grid, tmp_theta(inverse_mu(mu_grid)), 1, 
                                model.params['omega_A'], model.params['sigma_A'], model.params['omega_B'])
    marginal_product_worker = Fl(inverse_mu(mu_grid), mu_grid, tmp_theta(inverse_mu(mu_grid)), 1,
                                 model.params['omega_A'], model.params['sigma_A'], model.params['omega_B'])
    ax[1].plot(tmp_mu(xgrid), marginal_product_skill, color=colors[25 + i])
    
    # update the initial coefficients
    init_coefs = model.solver.coefs  
    
    print 'Done with omega_A = %g' % val

##### omega_A in (0.5, 0.1) #####

# initial guess for theta(x) is constant function
init_theta_coefs    = np.zeros(deg + 1)
init_theta_coefs[0] = model.workers.size / model.firms.size
        
# initial guess for mu(x) is identity function
y_lower, y_upper = model.firms.lower_bound, model.firms.upper_bound
init_mu_coefs = np.zeros(deg + 1)
init_mu_coefs[0] = 0.5 * (y_upper + y_lower)
init_mu_coefs[1] = 0.5 * (y_upper - y_lower)
        
# initial set of coefficients should have shape (2, deg + 1)
init_coefs = np.vstack((init_theta_coefs, init_mu_coefs))
tmp_coefs = init_coefs

param_values = np.linspace(0.5, 0.1, 25)
for i, val in enumerate(param_values):
    
    # change the parameter
    model.update_parameter('omega_A', val)
    
    # compute the equilibrium using collocation
    model.set_solver('OrthogonalCollocation', kind='Chebyshev')
    model.solver.solve(init_coefs, bc=boundary_conditions, matching='pam', method='hybr')
    
    tmp_theta = model.equilibrium['theta']
    tmp_mu    = model.equilibrium['mu']
    xgrid     = np.linspace(x_lower, x_upper, 1000)
    
    # mu(x) may be increasing (PAM) or decreasing (NAM)!
    data = np.hstack((xgrid[:,np.newaxis], tmp_mu(xgrid)[:,np.newaxis]))
    data = data[data[:,1].argsort()]
    inverse_mu = interpolate.PchipInterpolator(data[:,1], data[:,0])
        
    # grid of values for plotting the matching function
    mu_lower = tmp_mu(x_lower)
    mu_upper = tmp_mu(x_upper)
    mu_grid  = np.linspace(mu_lower, mu_upper, 1000)    

    marginal_product_skill = Fx(inverse_mu(mu_grid), mu_grid, tmp_theta(inverse_mu(mu_grid)), 1, 
                                model.params['omega_A'], model.params['sigma_A'], model.params['omega_B'])
    marginal_product_worker = Fl(inverse_mu(mu_grid), mu_grid, tmp_theta(inverse_mu(mu_grid)), 1,
                                 model.params['omega_A'], model.params['sigma_A'], model.params['omega_B'])
    ax[0].plot(tmp_mu(xgrid), marginal_product_skill, color=colors[25 - i])
    
    # update the initial coefficients
    init_coefs = model.solver.coefs  
    
    print 'Done with omega_A = %g' % val

# demarcate the schedule of points where effects balance
ax[0].axhline(0, ls='dashed', color='k')
ax[1].axhline(0, ls='dashed', color='k')

# axes, labels, title, colobar, etc
ax[0].set_xlabel('Firm productivity, $y$', fontsize=15, family='serif')
ax[1].set_xlabel('Firm productivity, $y$', fontsize=15, family='serif')

ax[0].set_xlim(0, y_upper)
ax[1].set_xlim(0, y_upper)

ax[0].set_yscale('log')
ax[1].set_yscale('log')
ax[0].set_ylabel(r'$\frac{F_x}{F_l}$',rotation='horizontal', fontsize=25)
ax[0].yaxis.set_label_coords(-0.15, 0.5)
#fig.suptitle('Varying $\omega_A$:\n"Skill-biased" or "quantity-biased" change?', 
#             x=0.5, y=1.05, fontsize=20, family='serif')
ax[0].grid()
ax[1].grid()
cax = fig.add_axes([0.01, 0.01, 0.99, 0.05])
cb = mpl.colorbar.ColorbarBase(cax, cmap=mpl.cm.jet, norm=mpl.colors.Normalize(vmin=0.05, vmax=0.95),
                               orientation='horizontal')
cb.set_label('$\omega_A$', fontsize=25)

#plt.savefig('graphics/figure-5-skill-vs-quantity-biased-change-omega-A.pdf', 
#            bbox_inches='tight')
plt.show()
```
