```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sym
from scipy import stats

import inputs
import models
import solvers
import pandas

from __future__ import division
```

```python
# define some workers skill
x, a, b = sym.var('x, a, b')
skill_cdf = (x - a) / (b - a)
skill_params = {'a': 0.9999, 'b': 1.0001}
skill_bounds = [skill_params['a'], skill_params['b']]

workers = inputs.Input(var=x,
                       cdf=skill_cdf,
                       params=skill_params,
                       bounds=skill_bounds,
                       )

# define some firms lognorm.pdf(x, s) = 1 / (s*x*sqrt(2*pi)) * exp(-1/2*(log(x)/s)**2)
y, mu2, sigma2 = sym.var('y, mu2, sigma2')
productivity_cdf = 0.5 + 0.5 * sym.erf((sym.log(y) - mu2) / (sym.sqrt(2) * sigma2))
productivity_params = {'mu2': -1.83, 'sigma2': 4.66}
productivity_bounds = [1e-3, 100.0]

firms = inputs.Input(var=y,
                     cdf=productivity_cdf,
                     params=productivity_params,
                     bounds=productivity_bounds,
                     )
```

```python
# Developped country

R, l, r, rho, gamma, eta, L, A, kapa = sym.var('R, l, r, rho, gamma, eta, L, A, kapa')

# First set k to root 3
k_star = (2*6**(-0.5)*(A*eta**2*kapa*x**0.5/R)**0.5*sym.cos(0.333333333333333*sym.acos(3*6**0.5*x**(-0.5)*(l*x*y/r)**0.25*(R*x**(-0.5)/(A*eta**2*kapa))**0.5*(-eta + 1)/(2*eta))))**(4)

F = A*kapa*(eta*(k_star*x)**rho + (1- eta)*(y*(l/r))**rho)**(gamma/rho)

F_params = {'R':0.13099, 'rho':0.25, 'gamma':0.5, 'eta':0.89, 'L':4.2, 'A': 1.0, 'kapa':1.0}
```

```python
# Try out positive
model = models.Model('positive',
                     workers=workers,
                     firms=firms,
                     production=F,
                     params=F_params)

solver = solvers.ShootingSolver(model=model)
```

```python
solver.solve(10, tol=1e-6, number_knots=6000, integrator='lsoda',
             with_jacobian=True, atol=1e-12, rtol=1e-9, check=False)
```

```python
solver.solution
```

```python
xs = []
thetas = []
ys = []
ws = []
pis = []
for i in range(5971):
    xs.append(solver.solution.index[i])
    ys.append(solver.solution.iloc[i,0])
    thetas.append(solver.solution.iloc[i,1])
    ws.append(solver.solution.iloc[i,2])
    pis.append(solver.solution.iloc[i,3])
    
kk = sym.lambdify((x,y,l,r,A,kapa,eta,R), k_star)
ks = []
for a in range(len(xs)):
    ks.append(kk(xs[a],ys[a],thetas[a],1,1,1,0.89,0.13099))
```

```python
plt.figure(1, figsize=(16, 4))
plt.suptitle("Rich Country, no variation in x", fontsize = 15)

plt.subplot(131)
plt.xlabel("$y$", fontsize = 20)
plt.ylabel("$k^{*},x$", fontsize=20)
plt.plot(ys, xs, label="Land quality")
plt.legend(loc=2)
plt.xlim(0,0.1)
plt.ylim(0,2)

plt.subplot(132)
plt.plot(ys, thetas, label="Labour", c='r')
plt.xlabel("$y$", fontsize = 20)
plt.legend(loc=2)
plt.xlim(0,0.1)
#plt.ylim(15,18)

plt.subplot(133)
plt.plot(ys, ks, label="Capital", c='g')
plt.xlabel("$y$", fontsize = 20)
plt.legend(loc=2)
plt.xlim(0,0.1)
plt.ylim(8,10)

plt.show()
```
